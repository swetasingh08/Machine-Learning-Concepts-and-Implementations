```python
# Implement and demonstrate the FIND-S algorithm for finding the most specific hypothesis based on a given set of training data samples. Read the training data from a .CSV file.
```

```python
import pandas as pd
import numpy as np

# Read CSV file
data = pd.read_csv("data_set.csv")
print("Dataset:\n")
print(data)

# Attributes (all columns except last)
d = np.array(data)[:, :-1]
print("\nThe attributes are:\n", d)

# Target (last column)
target = np.array(data)[:, -1]
print("\nThe target is:\n", target)

# Find-S Algorithm with steps
def train(c, t):
    
    # Step 1: Initialize with first positive example
    for i, val in enumerate(t):
        if val == "Yes":
            specific_hypothesis = c[i].copy()
            print("\nInitial hypothesis (First Positive Example):")
            print(specific_hypothesis)
            break

    # Step 2: Iterate through all examples
    for i, val in enumerate(c):
        print(f"\nStep {i+1}:")
        print("Training instance:", val, "Target:", t[i])

        if t[i] == "Yes":
            for x in range(len(specific_hypothesis)):
                if val[x] != specific_hypothesis[x]:
                    specific_hypothesis[x] = '?'

            print("Updated hypothesis:", specific_hypothesis)
        else:
            print("Negative example → No change")

    return specific_hypothesis

# Final hypothesis
final_h = train(d, target)
print("\nFinal Maximally Specific Hypothesis:\n", final_h)
```
