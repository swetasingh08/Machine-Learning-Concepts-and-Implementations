```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import pandas as pd

# -----------------------------
# Load Dataset
# -----------------------------
news = fetch_20newsgroups()

print("\n" + "="*50)
print("ALL AVAILABLE CATEGORIES")
print("="*50)

for i, category in enumerate(news["target_names"]):
    print(f"{i+1:2d}. {category}")

# -----------------------------
# Select Categories
# -----------------------------
categories = [
    'alt.atheism',
    'soc.religion.christian',
    'comp.graphics',
    'sci.med'
]

news_train = fetch_20newsgroups(
    subset='train', categories=categories, shuffle=True)

news_test = fetch_20newsgroups(
    subset='test', categories=categories, shuffle=True)

print("\n" + "="*50)
print("SELECTED CATEGORIES")
print("="*50)

for cat in news_train.target_names:
    print(f"- {cat}")

# -----------------------------
# Build Model
# -----------------------------
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('nb', MultinomialNB())
])

model.fit(news_train.data, news_train.target)

# -----------------------------
# Prediction
# -----------------------------
predicted = model.predict(news_test.data)

# -----------------------------
# Results
# -----------------------------
print("\n" + "="*50)
print("MODEL PERFORMANCE")
print("="*50)

accuracy = metrics.accuracy_score(news_test.target, predicted)
print(f"Accuracy : {accuracy:.4f} ({accuracy*100:.2f}%)")

# -----------------------------
# Classification Report (Compact)
# -----------------------------
print("\nClassification Report:")
report = metrics.classification_report(
    news_test.target,
    predicted,
    target_names=news_test.target_names,
    digits=3
)
print(report)

# -----------------------------
# Confusion Matrix (Table Format)
# -----------------------------
print("Confusion Matrix:")

cm = metrics.confusion_matrix(news_test.target, predicted)

df_cm = pd.DataFrame(
    cm,
    index=news_test.target_names,
    columns=news_test.target_names
)

print(df_cm.to_string())

print("\n" + "="*50)
```
